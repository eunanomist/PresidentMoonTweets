```python
import pandas as pd
import os
import numpy as np
```

```python
# working directory check
print(os.getcwd())
```

```python
#read data
df = pd.read_csv("Moon.csv", engine = 'python')
```

```python
df
```

```python
# try to remove columns
df_1 = df.drop([df.columns[0],df.columns[2],df.columns[-1],df.columns[-2],df.columns[-3]], axis =1)
```

```python
df_1
```

```python
df_1 = df_1[pd.notnull(df_1.text)]
```

```python
df_1
```

```python
df_1.to_csv(r'C:\Users\Sophie\Desktop\df_1.csv',encoding='utf-8-sig')
# Check 해보니 중간에 조금 비어 있음  일단 데이터 합치고 날리겠음
```

```python
grouped = df_1.groupby('tweet_date')
weighted_avg_func = lambda g: np.average(g['favorites'], weights = None)
avg_fav = pd.DataFrame(grouped.apply(weighted_avg_func))
avg_fav.columns = ['avg_fav']
avg_fav
```

```python
grouped = df_1.groupby('tweet_date')
weighted_avg_func = lambda g: np.average(g['retweets'], weights = None)
avg_ret = pd.DataFrame(grouped.apply(weighted_avg_func))
avg_ret.columns = ['avg_ret']
avg_ret
```

```python
# try to grouping by data
pro_text = df_1[[ df_1.columns[0], df_1.columns[1]]].groupby(df_1.columns[0]).sum().sort_values(by=df_1.columns[0])
pro_text
```

```python
# Mid Check
df_merge = pd.concat([pro_text, avg_ret], axis=1, sort= False)
df_mid = pd.concat([df_merge, avg_fav], axis=1, sort= False)
df_mid
```

```python
df_mid.to_csv(r'C:\Users\Sophie\Desktop\df_mid.csv',encoding='utf-8-sig')
# Check 해보니 중간에 조금 비어 있음  일단 데이터 합치고 날리겠음
```

```python
#read data
df_mid2 = pd.read_csv("df_mid.csv", engine = 'python', encoding='utf-8-sig')
df_mid2
```

```python
# Outofrange 날리기
df_mid2 = df_mid2.drop(index=0)
```

```python
df_mid2['tweet_date'] = pd.to_datetime(df_mid2['tweet_date'])
```

```python
df_mid2
```

```python
df_mid2.isnull().values.any()
```

```python
df_mid5 = df_mid2.groupby(pd.Grouper(key = df_mid2.columns[0], freq = '7D'))['text'].apply(lambda x: " ".join(x.astype(str)))
df_text = pd.DataFrame(data = df_mid5) 
df_text
```

```python
df_text.isnull().values.any()
df_text.to_csv(r'C:\Users\Sophie\Desktop\df_text.csv',encoding='utf-8-sig')
# Check 해보니 중간에 조금 비어 있음  일단 데이터 합치고 날리겠음
```

```python
df_ret = df_mid2.groupby(pd.Grouper(key = df_mid2.columns[0], freq = '7D'))['avg_ret'].mean()
df_ret = pd.DataFrame(data = df_ret) 
df_ret
```

```python
df_ret.isnull().values.any()
```

```python
df_ret.to_csv(r'C:\Users\Sophie\Desktop\df_ret.csv',encoding='utf-8-sig')
# Check 해보니 중간에 조금 비어 있음  일단 데이터 합치고 날리겠음
```

```python
df_fav = df_mid2.groupby(pd.Grouper(key = df_mid2.columns[0], freq = '7D'))['avg_fav'].mean()
df_fav = pd.DataFrame(data = df_fav) 
df_fav
```

```python
df_fav.isnull().values.any()
```

```python
grouped = df_fav.groupby('tweet_date')
weighted_avg_func = lambda g: np.average(g['avg_fav'], weights = None)
df_fav = pd.DataFrame(grouped.apply(weighted_avg_func))
df_fav.columns = ['favorite']
df_fav
```
